```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file=pd.read_csv(r'E:\LST project\manuscript\new\LST_ALL_results 15March21.csv')
```

```python
file
```

```python
x=file['Ground_Temperature']
SMWA_1=file['Ermida_LST']
SMWA_2=file['SMWA_LST1']
SMWA_3=file['SMWA_LST2']
USGS=file['USGS_LST']
GSCA1=file['SCA_LST1']
GSCA2=file['SCA_LST2']
```

```python
correlation_matrix = np.corrcoef(x, SMWA_1)
correlation_xy = correlation_matrix[0,1]
r_squared1 = round(correlation_xy**2,3)
print('r_squared 1=', r_squared1)

correlation_matrix = np.corrcoef(x, SMWA_2)
correlation_xy = correlation_matrix[0,1]
r_squared2 = round(correlation_xy**2,3)
print('r_squared 2=', r_squared2)

correlation_matrix = np.corrcoef(x, SMWA_3)
correlation_xy = correlation_matrix[0,1]
r_squared3 = round(correlation_xy**2,3)
print('r_squared 3=', r_squared3)

correlation_matrix = np.corrcoef(x, USGS)
correlation_xy = correlation_matrix[0,1]
r_squared4 = round(correlation_xy**2,3)
print('r_squared 4=', r_squared4)

correlation_matrix = np.corrcoef(x, GSCA1)
correlation_xy = correlation_matrix[0,1]
r_squared5 = round(correlation_xy**2,3)
print('r_squared 5=', r_squared5)

correlation_matrix = np.corrcoef(x, GSCA2)
correlation_xy = correlation_matrix[0,1]
r_squared6 = round(correlation_xy**2,3)
print('r_squared 6=', r_squared6)

```

```python
fig,((axs1,axs2,axs3),(axs4,axs5,axs6))=plt.subplots(2,3,figsize=(15,10))
axs1.scatter(x,SMWA_1)
axs1.plot(x,x)
# axs1.set_xlabel('Insitu LST(K)',fontsize =11)
axs1.set_ylabel('SMWA_A LST (K)',fontsize =14)
axs1.text(288,332,"(a)",fontsize=13)
axs1.text(310,296,"RMSE = 1.868 K",fontsize=13)
axs1.text(310,292,"$R^2$ = "+str(r_squared1),fontsize=13)
axs1.text(310,289,"$N$ = 89",fontsize=13)

axs2.scatter(x,SMWA_2)
axs2.plot(x,x)
axs2.set_xlabel('In situ LST (K)',fontsize =14,labelpad=18)
axs2.set_ylabel('SMWA_C LST (K)',fontsize =14)
axs2.text(288,332,"(b)",fontsize=13)
axs2.text(310,296,"RMSE = 1.703 K",fontsize=13)
axs2.text(310,292,"$R^2$ = "+str(r_squared2),fontsize=13)
axs2.text(310,289,"$N$ = 89",fontsize=13)

axs3.scatter(x,SMWA_3)
axs3.plot(x,x)
# axs3.set_xlabel('Insitu LST(K)',fontsize =11)
axs3.set_ylabel('SMWA_N LST (K)',fontsize =14)
axs3.text(288,332,"(c)",fontsize=13)
axs3.text(310,296,"RMSE = 1.725 K",fontsize=13)
axs3.text(310,292,"$R^2$ = "+str(r_squared3),fontsize=13)
axs3.text(310,289,"$N$ = 89",fontsize=13)

axs4.scatter(x,USGS)
axs4.plot(x,x)
# axs4.set_xlabel('Insitu LST(K)',fontsize =11)
axs4.set_ylabel('Landsat level-2 LST (K)',fontsize =14)
axs4.text(288,332,"(d)",fontsize=13)
axs4.text(310,296,"RMSE = 3.524 K",fontsize=13)
axs4.text(310,292,"$R^2$ = "+str(r_squared4),fontsize=13)
axs4.text(310,289,"$N$ = 89",fontsize=13)

axs5.scatter(x,GSCA1)
axs5.plot(x,x)
axs5.set_xlabel('In situ LST (K)',fontsize =14,labelpad=18)
axs5.set_ylabel('GSCA_C LST (K)',fontsize =14)
axs5.text(288,332,"(e)",fontsize=13)
axs5.text(310,296,"RMSE = 2.620 K",fontsize=13)
axs5.text(310,292,"$R^2$ = "+str(r_squared5),fontsize=13)
axs5.text(310,289,"$N$ = 89",fontsize=13)

axs6.scatter(x,GSCA2)
axs6.plot(x,x)
# axs6.set_xlabel('Insitu LST(K)',fontsize =11)
axs6.set_ylabel('GSCA_N LST (K)',fontsize =14)
axs6.text(288,332,"(f)",fontsize=13)
axs6.text(310,296,"RMSE = 2.600 K",fontsize=13)
axs6.text(310,292,"$R^2$ = "+str(r_squared6),fontsize=13)
axs6.text(310,289,"$N$ = 89",fontsize=13)
fig.tight_layout()
fig.savefig(r'E:\LST project\manuscript\new\LST_scatter_new.jpeg',dpi=700)
```
